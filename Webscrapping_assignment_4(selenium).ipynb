# Q1: Scrape the details of most viewed videos on YouTube from Wikipedia: URl=https://en.wikipedia.org/wiki/Listofmost-viewedYouTubevideos/. Find the folowing details:
    A) Rank
    B) Name
    C) Artist
    D) Upload date
    E) Views

```python
url="https://en.wikipedia.org/wiki/List_of_most-viewed_YouTube_videos" #passing url
driver=webdriver.Chrome(r"C:\Users\Win 10\Downloads\chromedriver_win32\chromedriver")
driver=webdriver.Chrome("chromedriver")
driver.get(url)
time.sleep(2)

rank=[]             #empty list 
for i in range(1):
    for j in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[1]"):#fetching rank tag using xpath
        rank.append(j.text.replace('.',""))
name=[]
for i in range(1):
    for j in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[2]/a"): #fetching name tag using xpath
        name.append(j.text)

artist=[]
for i in range(1):
    for j in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[3]/a"):# fetching artists tag using xpath
        artist.append(j.text)

uplddate=[]
for i in range(1):
    for j in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[5]"): #fetching upload date tag using xpath
        uplddate.append(j.text)

views=[]
for i in range(1):
    for j in driver.find_elements_by_xpath("//*[@id='mw-content-text']/div[1]/table[3]/tbody/tr/td[4]"): #fetching views tag using xpath
        views.append(j.text)


print("Rank:",rank)
print("\n","Name:",name)
print("\n","Artist:",artist)
print("\n","Upload date:",uplddate)                    #printing the outputs
print("\n","Views(billions):",views)
driver.close()
```
